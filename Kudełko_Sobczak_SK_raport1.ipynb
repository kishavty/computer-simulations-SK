## Rozkład Laplace'a
Jest to ciągły rozkład prawdopodobieństwa. Nazywany także "dwustronnym rozkładem wykładniczym", ze względu na jego powstawanie - mianowicie powstaje on podczas odejmowania dwóch rozkładów wykładniczych.

Jeśli $X_1$ i $X_2$ są niezależnymi zmiennymi losowymi o rozkładzie wykładniczym z parametrem $\lambda = 1/b$ to zmienna losowa:
$\mu = X_1 - X_2$ ma rozkład Laplace'a o średniej $\mu$ i czynniki skali $b$.  


* Dystrybuanta:
\begin{equation*}
  F(x) =
  \begin{cases}
    \frac{1}{2}e^{-\frac{|x - \mu|}{b}}   & \quad \text{dla } x \leq \mu \\
    1 - \frac{1}{2}e^{-\frac{|x - \mu|}{b}} & \quad \text{dla } x \geq \mu
  \end{cases}
\end{equation*}
* Gęstość:
\begin{equation*}
f(x) = \frac{1}{2b}e^{-\frac{|x - \mu|}{b}}
\end{equation*}
* Funkcja charakterystyczna:
\begin{equation*}
  \varphi(t) = \frac{e^{\mu t}}{1 - b^{2}t^{2}}
\end{equation*}
* Wartość oczekiwana:
\begin{equation*}
  \mu
\end{equation*}
* Mediana:
\begin{equation*}
  \mu
\end{equation*}
* Moda:
\begin{equation*}
  \mu
\end{equation*}
* Wariancja:
\begin{equation*}
  2b^{2}
\end{equation*}
* Współczynnik skośności:
\begin{equation*}
  0
\end{equation*}
* Kurtoza:
\begin{equation*}
  3
\end{equation*}

## Rozkład Weibulla
Jest to ciągły rozkład prawdopodobieństwa, który spełnia potęgowe prawo skalowania. Poza ekonomią nazywany rozkładem "Bradforda". Może on w zależności od parametrów przypominać zarówno rozkład normalny jak i wykładniczy.
* Dystrybuanta:
\begin{equation*}
F(x) = 1 - e^{-(\frac{x}{\lambda})^k}
\end{equation*}
* Gęstość:
\begin{equation*}
f(x) = \frac{k}{\lambda}(\frac{x}{\lambda})^{k-1}e^{-(\frac{x}{\lambda})^k}
\end{equation*}
* Funkcja charakterystyczna:
Nie ma na ogół jawnej postaci.
* Wartość oczekiwana:
\begin{equation*}
  \lambda \Gamma(1 + \frac{1}{k})
\end{equation*}
* Mediana:
\begin{equation*}
  \lambda ln(2)^{\frac{1}{k}}
\end{equation*}
* Moda:
\begin{equation*}
  \lambda(\frac{k - 1}{k})^{\frac{1}{k}}
\end{equation*}
* Wariancja:
\begin{equation*}
  \lambda^{2} \Gamma(1 + \frac{2}{k}) - \mu^{2}
\end{equation*}
* Współczynnik skośności:
\begin{equation*}
  \Gamma(1 + \frac{3}{k}) \lambda^{3} - 3\mu\sigma^{2} - \mu^{3}
\end{equation*}
* Kurtoza:
\begin{equation*}
  \frac{-6 \Gamma _1^{4}+ 12\Gamma _1^{2}\Gamma _2-3\Gamma_ 2^{2}- 4\Gamma_ 1 \Gamma_ 3 + \Gamma_ 4}{[\Gamma_ 2 - \Gamma_ 1^{2}]^{2}}
\end{equation*}

## Rozkład wykładniczy
Jest to ciągły rozkład prawdopodobieństwa, który opisuje odstępy czasu pomiędzy kolejnymi zdarzeniami, gdy w jednostce czasu zachodzi średnio $\lambda$ niezależnych zdarzeń. Posiada własność tzw. *braku pamięci*. Charakteryzuje się tym, że prawdopodobieństwo wystąpienia zjawiska w danym przedziale czasowym zależy jedynie od długości tego przedziału i jest niezależne od tego, czy zjawisko wystąpiło w jakimkolwiek innym przedziale czasowym: $\mathbb {P} \left(X>s+t|X>s\right)=\mathbb {P} \left(X>t\right)$. Rozkład ten jest przypadkiem szczególnym rozkładu Gamma, tj. gdy $ X \sim \Gamma(1,\lambda)$, to $X \sim Exp(\lambda).$
* Dystrybuanta:
\begin{equation*}
F(x) = 1-e^{-\lambda x}
\end{equation*}
* Gęstość:
\begin{equation*}
f(x) = \lambda e^{-\lambda x}\cdot \mathbf {1} _{[0,\infty )}
\end{equation*}
* Funkcja charakterystyczna:
\begin{equation*}  
\varphi(t)= \left(1-{\frac {it}{\lambda }}\right)^{-1}
\end{equation*}
* Wartość oczekiwana:
\begin{equation*}
  {\frac {1}{\lambda }}
\end{equation*}
* Mediana:
\begin{equation*}
  {\frac {\ln(2)}{\lambda }}
\end{equation*}
* Moda:
\begin{equation*}
  0
\end{equation*}
* Wariancja:
\begin{equation*}
  {\frac {1}{\lambda ^{2}}}
\end{equation*}
* Współczynnik skośności:
\begin{equation*}
  2
\end{equation*}
* Kurtoza:
\begin{equation*}

\end{equation*}

# Metoda odwracania dystrybuanty:


## Rozkład ciągły
Niech $X$ będzie teraz ciągłą zmienną losową o ściśle rosnącej dystrybuancie $F_{X}(x)$. Wtedy: \
$F_{X}^{-1}(U) = X$, gdzie $U$ ~ $U(0,1)$.



### Algorytm
1. Generuj $U$ ~ $U(0,1)$.
2. Wstaw $X = F_{X}^{-1}(U)$.

## Rozkład Laplace'a - przykład I

### Implementacja

```python
import numpy as np
from scipy.stats import laplace
import matplotlib.pyplot as plt
import scipy.stats as st
import statsmodels.api as sm

def laplace_inv_cdf(y, mu, b):
    if y < 0.5:
        return mu + b * np.log(2 * y)
    else:
        return mu - b * np.log(2 - 2 * y)

def laplace_inv_cdf_random_array(mu, b, random_ys_count):
    random_ys = np.random.uniform(size = random_ys_count)
    output_xs = []
    for y in random_ys:
        output_xs.append(laplace_inv_cdf(y, mu, b))
    return output_xs
```

### Testy


```python
### WYGENEROWANIE PRÓBY
mu = 0
b = 1
n = 1000
sample = laplace_inv_cdf_random_array(mu, b, n)
```

```python
### GESTOSC TEORETYCZNA I EMPIRYCZNA
hist, edges = np.histogram(sample, bins = 30, density=True)
x_hist = (edges[1:] + edges[:-1]) / 2

theoretical_x = np.linspace(mu - 4 * b, mu + 4 * b, 1000)
theoretical_laplace = laplace.pdf(theoretical_x, loc=mu, scale=b)

plt.plot(theoretical_x, theoretical_laplace, label='Theoretical density')
plt.bar(x_hist, hist, width=(edges[1] - edges[0]), alpha=0.5, label='Empirical density')
plt.ylabel('Values')
plt.xlabel('X')
plt.title("Laplace density and histogram")
plt.legend()
plt.show()
```

```python
### DYSTRYBUANTA TEORETYCZNA I EMPIRYCZNA
sorted_sample = np.sort(sample)
x = np.linspace(sorted_sample.min(), sorted_sample.max(), 1000)
cdf_empirical = np.array([(sorted_sample <= xi).mean() for xi in x])
cdf_theoretical = st.laplace.cdf(x)
plt.plot(x, cdf_empirical, label='Empirical distribution', color='g')
plt.plot(x, cdf_theoretical, label='Theoretical distribution', color='b')
plt.ylabel('Values')
plt.title("Laplace distribution")
plt.xlabel('X')
plt.legend()
plt.show()
```

```python
### WYKRES KWANTYLOWY (QQ-PLOT)
plt.figure()
st.probplot(sample, dist=st.laplace, sparams=(mu,b), plot=plt)
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered values')
plt.title('QQ-plot Laplace')
plt.show()
```

```python
### ŚREDNIA I WARIANCJA
theoretical_mean = mu
theoretical_variance = 2 * b**2

empirical_mean = np.mean(sample)
empirical_variance = np.var(sample)

print(f"Theoretical mean: {theoretical_mean}")
print(f"Empirical mean: {empirical_mean}")
print(f"Theoretical variance: {theoretical_variance}")
print(f"Empirical variance: {empirical_variance}")
```

## Rozkład Weibulla - przykład II

### Implementacja

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import math

def weibull_inv_cdf(y, k, l):
    return l * (-np.log(1 - y))**(1 / k)

def weibull_inv_cdf_random_array(l, k, random_ys_count):
    random_ys = np.random.uniform(size = random_ys_count)
    output_xs = []
    for y in random_ys:
        output_xs.append(weibull_inv_cdf(y, l, k))
    return output_xs
```

### Testy

```python
### WYGENEROWANIE PRÓBY
k = 2
l = 3
n = 10000
sample = weibull_inv_cdf_random_array(k, l, n)
```

```python
### GESTOSC TEORETYCZNA I EMPIRYCZNA

theoretical_x = np.linspace(0, 20, 1000)
theoretical_weibull = (k / l) * (theoretical_x / l)**(k - 1) * np.exp(-(theoretical_x / l)**k)
plt.figure(figsize=(10, 5))
plt.hist(sample, bins=50, density=True, label='Empirical density')
plt.plot(theoretical_x, theoretical_weibull, label='Theoretical density')
plt.xlabel('x')
plt.title("Weibull density and histogram")
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
### DYSTRYBUANTA TEORETYCZNA I EMPIRYCZNA
sorted_sample = np.sort(sample)
x = np.linspace(sorted_sample.min(), sorted_sample.max(), 1000)
cdf_empirical = np.array([(sorted_sample <= xi).mean() for xi in x])
cdf_theoretical = st.weibull_min.cdf(x, c=k, scale=l)
plt.plot(x, cdf_empirical, label='Empirical distribution', color='g')
plt.plot(x, cdf_theoretical, label='Theoretical distribution', color='b')
plt.title("Weibull distribution")
plt.ylabel('Values')
plt.xlabel('X')
plt.legend()
plt.show()

```

```python
### WYKRES KWANTYLOWY (QQ-PLOT)
plt.figure()
st.probplot(sample, dist=st.exponweib, sparams=(l,k), plot=plt)
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered values')
plt.title('QQ-plot Weibull ')
plt.show()
```

```python
### ŚREDNIA I WARIANCJA
theoretical_mean = l * (np.log(2))**(1 / k)
theoretical_variance = (l**2) * (math.gamma(1 + 2 / k) - (math.gamma(1 + 1 / k))**2)

empirical_mean = np.mean(sample)
empirical_variance = np.var(sample)

print(f"Theoretical mean: {theoretical_mean}")
print(f"Empirical mean: {empirical_mean}")
print(f"Theoretical variance: {theoretical_variance}")
print(f"Empirical variance: {empirical_variance}")
```

## Rozkład dyskretny
Niech $X$ będzie zmienną losową o rozkładzie: \
$P(X = x_i) = p_i, i = 1, . . . , n$, gdzie $ \sum_{i =1}^{n} p_i = 1$.


### Algorytm
1. Generuj $U$ ~ $U(0,1)$.
2. Jeżeli $ \sum_{i = 1}^{j - 1}p_i < U \leq \sum_{i = 1}^{j}p_i$, dla dowolnego $j \in {1,2,...,n}$, to wstaw $X = x_j$.

## Rozkład jednostajny dyskretny

### Implementacja


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.stats as st


def discrete_uniform_gen(a, b, num):
    U_list = np.random.uniform(0,1,size=num)
    X_list = np.zeros(num)
    for i, u in enumerate(U_list):
        X_list[i] = np.ceil((b-a+1)*u -1 + a)
    return X_list
```

### Testy

```python
### WYGENEROWANIE PRÓBY
a = 1
b = 6
n = b - a + 1
num = 10000

sample = discrete_uniform_gen(a, b, num)
```

```python
### GĘSTOŚĆ TEORETYCZNA I EMPIRYCZNA
def uniform_discrete_pdf(k, a, b):
    if a <= k and k <= b:
        return 1 / n
    else:
        return 0

plt.plot(np.arange(a, b+1), np.ones(n) / n, 'ro-', label="Empirical density")
plt.hist(sample, bins=np.arange(a-0.5, b+1.5), density=True, alpha=0.5, label="Theoretical density")
plt.xlabel('x')
plt.ylabel('Values')
plt.title("Uniform discrete density and histogram")
plt.legend()
plt.show()
```

```python
### DYSTRYBUANTA TEORETYCZNA I EMPIRYCZNA
def uniform_discrete_cdf(x, a, b):
    if x < a:
        return 0
    elif x > b:
        return 1
    else:
        return (x - a + 1) / (b - a + 1)

plt.figure()
plt.step(np.sort(sample), [uniform_discrete_cdf(x, a, b) - (1/n) for x in np.sort(sample)], label='theoretical distribution')
plt.plot(np.sort(sample), np.arange(num) / num, label='empirical distribution', alpha=0.5)
plt.xlabel("x")
plt.ylabel("Values")
plt.title("Uniform discrete distribution")
plt.legend()
plt.show()
```

```python
### WYKRES KWANTYLOWY (QQ-PLOT)
sorted_sample = np.sort(sample)
plt.figure()
st.probplot(sorted_sample, plot=plt, dist=scipy.stats.randint, sparams=(a,b))
plt.show()
```

```python
### ŚREDNIA I WARIANCJA
theoretical_mean = (a + b) / 2
theoretical_variance = ((b - a + 1)**2 - 1) / 12

empirical_mean = np.mean(sample)
empirical_variance = np.var(sample)

print(f"Theoretical mean: {theoretical_mean}")
print(f"Empirical mean: {empirical_mean}")
print(f"Theoretical variance: {theoretical_variance}")
print(f"Empirical variance: {empirical_variance}")
```

# Metoda akceptacji- odrzucenia
## Rozkład ciągły
### Cel:
Chcemy generować realizację zmiennej losowej $X$, o gęstości $f(x)$ =
$P(X = i) = p_i, (i = 1,2,...n).$ \
### Założenia:
1. Potrafimy efektywnie generować inną zmienną losową $Y$ o gęstości $g(y)$, która przyjmuje wartości z tego samego zbioru $A$ co zmienna losowa $X$,
2. Potrafimy wyznaczyć stałą $ c > 0$, spełniającą
\begin{equation*}
\sup_{x \in A} \frac{f(x)}{g(x)} \leq c
\end{equation*}\
Gdy mamy ”=”, to stała c jest wyznaczona optymalnie.
### Algorytm:
1. Generuj $Y$.
2. Generuj $U$ ~ $U(0,1)$, $\: U \perp y$.
3. Jeśli $U \leq \frac{f(Y)}{cg(Y)}$,  wstaw $X = Y$. W przeciwnym razie wróć do punktu 1.

## Rozkład Laplace'a - przykład I

### Implementacja

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

def laplace_generator(x, mu, b):
    return (1 / (2 * b)) * np.exp(-np.abs(x - mu) / b)


def laplace_rejection_method(mu, b, size = 1):
    samples = []
    for i in range(size):
        U1 = np.random.uniform(-5,5)
        U2 = np.random.random()
        c = laplace_generator(U1, mu, b)
        if U2 <= c:
            samples.append(U1)

    else:
        return np.array(samples)


```

### Testy

```python
### WYGENEROWANIE PRÓBY
mu = 0
b = 1
sample = laplace_rejection_method(mu, b, size = 100000)
```

```python
### GESTOSC TEORETYCZNA I EMPIRYCZNA

theoretical_x = np.linspace(-20, 20, 1000)
theoretical_Laplace =  (1 / (2 * b)) * np.exp(-np.abs(theoretical_x - mu) / b)

plt.figure(figsize=(10, 5))
plt.hist(sample, bins = 30, density=True, label='Empirical density')
plt.plot(theoretical_x, theoretical_Laplace, label='Theoretical density')
plt.xlabel('x')
plt.ylabel('Values')
plt.title("Laplace density and histogram")
plt.legend()
plt.show()
```

```python
### DYSTRYBUANTA TEORETYCZNA I EMPIRYCZNA
sorted_sample = np.sort(sample)
x = np.linspace(sorted_sample.min(), sorted_sample.max(), 1000)
cdf_empirical = np.array([(sorted_sample <= xi).mean() for xi in x])
cdf_theoretical = st.laplace.cdf(x, scale = 1)
plt.plot(x, cdf_empirical, label='Empirical distribution', color='g')
plt.plot(x, cdf_theoretical, label='Theoretical distribution', color='b')
plt.ylabel('Values')
plt.xlabel('X')
plt.title("Laplace distribution")
plt.legend()
plt.show()
```

## Rozkład Weibulla - przykład II

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def weibull_generator(x, k, l):
    return (k / l) * (x / l)**(k - 1) * np.exp(-(x / l)**k)

def weibull_rejection_method(k, l, size=1):
    samples = []
    for i in range(size):
        U1 = np.random.uniform(-10,10)
        U2 = np.random.random()
        c = weibull_generator(U1, k, l)
        if U2 <= c:
            samples.append(U1)
    else:
        return np.array(samples)
```

## Rozkład dyskretny
### Cel:
Chcemy generować realizację zmiennej losowej $X$, o rozkładzie dyskretnym \
$P(X = i) = p_i, (i = 1,2,...n).$
### Założenie:
Potrafimy efektywnie generować inną zmienną losową $Y$ o rozkładzie \
$P(Y = i) = q_i, (i = 1,2,...n).$
### Algorytm:
1. Generujemy $Y$.
2. Generujemy $U$ ~ $U(0,1)$, $U,Y$ są niezależne.
3. Jeśli $U \leq \frac{p_Y}{cq_y}$, gdzie $c = \max_{i} \frac{p_i}{q_i}$, wstaw $X = Y$. W przeciwnym razie wróć do punktu 1.

## Rozkład jednostajny dyskretny

### Implementacja

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st

def uniform_discrete_rejection_method(a, b, num):
    sample = []
    while len(sample) < num:
        x = np.random.randint(a, b+1)
        u = np.random.uniform(0,1)
        if u <= 1 / (b - a + 1):
            sample.append(x)
    return np.array(sample)

```

### Testy

```python
### WYGENEROWANIE PRÓBY

a = 1
b = 6
num = 100000
sample = uniform_discrete_rejection_method(a,b,num)
n=b-a+1
```

```python
### DYSTRYBUANTA TEORETYCZNA I EMPIRYCZNA
def uniform_discrete_cdf(x, a, b):
    if x < a:
        return 0
    elif x > b:
        return 1
    else:
        return (x - a + 1) / (b - a + 1)



plt.figure()
plt.step(np.sort(sample), [uniform_discrete_cdf(x, a, b) - (1/n) for x in np.sort(sample)], label='theoretical distribution')
plt.plot(np.sort(sample), np.arange(num) / num, label='empirical distribution', alpha=0.5)
plt.xlabel("x")
plt.ylabel("Values")
plt.title("Uniform discrete distribution")
plt.legend()
plt.show()

```

## Metoda Boxa-Mullera

Jest to metoda generująca zmienne losowe $X, Y$ o rozkładzie normalnym, używająca dwóch zmiennych losowych z rozkładu jednostajnego na przedziale $[0,1]$. Zmienne $X$ i $Y$ otrzymane za pomocą tego algorytmu są niezależne i mają rozkład normalny o średniej równiej $0$ oraz wariancji równej $1$.

Wyprowadzenie tej metody opiera się na dwuwymiarowym układzie kartezjańskim, gdzie kąt $\alpha$ między poziomą osią a promieniem wodzącym punktu $(X,\:Y)$ ma rozkład $\alpha \sim U(0, 2\pi)$, natomiast $R^2 = X^2 + Y^2 $ ma rozkład $R^2 \sim Exp(\frac{1}{2}). \:$ [1]


### Algorytm

1. Generuj $U_1 \sim U(0,1), \: U_2 \sim U(0,1)$.
2. Wstaw $ X = \sqrt{ -2 \ln{U_1} } \cdot \cos{(2 \pi U_2) }, \: \: Y = \sqrt{ -2 \ln{U_1} } \cdot \sin{(2 \pi U_2) } $.


Wtedy $X \perp Y \:\: oraz \:\: X, Y \sim N(0,1)$.

### Implementacja

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st

def boxmuller_gen(mu=0, sigm=1, n=1000):
    """ Zwraca n par niezależnych zmiennych losowych X oraz Y z rozkładu N(mu,sigm). """
    samples_X, samples_Y = np.zeros(n), np.zeros(n)
    for i in range(n):
        U1, U2 = np.random.uniform(0,1,size=1), np.random.uniform(0,1,size=1)
        X, Y = sigm * (np.sqrt(-2*np.log(U1))*np.cos(2*np.pi*U2)) + mu , sigm * (np.sqrt(-2*np.log(U1))*np.sin(2*np.pi*U2)) + mu
        samples_X[i] = X
        samples_Y[i] = Y
    return samples_X, samples_Y
```

### Testy

```python
### WYGENEROWANIE PRÓBY
mu = 0
sigm = 1
n = 10000

boxmuller_values_X, boxmuller_values_Y = boxmuller_gen(mu, sigm, n)
```

```python
### WARTOŚĆ OCZEKIWANA, WARIANCJA, SKOŚNOŚĆ, KURTOZA
mean = np.mean(boxmuller_values_X)
variance = np.var(boxmuller_values_X)

print(f"Mean: {mean}") #oczekujemy wartosci bliskiej mu
print(f"Variance: {variance}") #oczekujemy wartosci bliskiej sigma**2
print(f"Skewness: {np.mean((boxmuller_values_X - mean)**3) / variance**(3/2)}") #oczekujemy wartosci bliskiej 0
print(f"Curtosis: {np.mean((boxmuller_values_X - mean)**4) / variance**2 - 3}") #oczekujemy wartosci bliskiej 0
```

```python
### GESTOSC TEORETYCZNA I EMPIRYCZNA
x_val = np.linspace(-4, 4, 1000)
plt.hist(boxmuller_values_X, bins = "sqrt", density = True, label="empirical density")
plt.plot(x_val, [(1/(sigm*(np.pi*2)**(0.5)))*np.exp((-(x-mu)**2)/(2*sigm**2)) for x in x_val], label="theoretical density")
plt.title("Histogram, mu={}, sigma={}".format(mu, sigm))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
### DYSTRYBUANTA TEORETYCZNA I EMPIRYCZNA
plt.figure()
plt.plot(np.sort(boxmuller_values_X), np.linspace(0, 1, n), label="Empirical distribution function")
plt.plot(np.linspace(-4*sigm+mu,4*sigm+mu,100), st.norm.cdf(np.linspace(-4*sigm+mu,4*sigm+mu,100), loc=mu, scale=sigm), alpha = 0.6, label="Theoretical distribution function")
plt.title("Distribution function, mu={}, sigma={}".format(mu, sigm))
plt.xlabel("x")
plt.ylabel("F(x)")
plt.legend()
plt.show()
```

```python
### WYKRES KWANTYLOWY (QQ-PLOT)
plt.figure()
st.probplot(np.sort(boxmuller_values_X), plot=plt)
plt.show()
```

## Metoda biegunowa

Ten algorytm, podobnie jak metoda Boxa-Mullera, generuje niezależne zmienne losowe z rozkładu normalnego. Działa jednak wydajniej niż poprzednia metoda, ponieważ zamiast bezpośredniego wyliczania funkcji $\sin{(2\pi U_2)}$ oraz $\cos{(2\pi U_2)}$, wykorzystuje pośrednie obliczenie sinusa i cosinusa losowego kąta. Niezależne zmienne losowe $V_1, V_2 \sim U(-1,1)$ to współrzędne punktu $(V_1, V_2)$, który jeśli należy do okręgu jednostkowego, to zostaje użyty do wygenerowania nowych niezależnych zmiennych losowych ze standardowego rozkładu normalnego.

### Algorytm

1. Generuj $V_1 \sim U(-1,1), \: V_2 \sim U(-1,1), \: V_1 \perp V_2$.
2. Wyznacz $ R^{2} = V_1^{2} + V_2^{2}.$
3. Jeśli $R^{2} > 1$, wróć do 1.
4. Wstaw $ X = \sqrt{\frac{-2ln{(R^{2})}}{R^{2}}} \cdot V_1, \:\: Y = \sqrt{\frac{-2ln{(R^{2})}}{R^{2}}} \cdot V_2$. \


Wtedy $X \perp Y \:\: oraz \:\: X, Y \sim N(0,1)$.

### Implementacja

```python
def polar_gen(mu=0, sigm=1, n=1000):
    """ Zwraca n par niezależnych zmiennych losowych X oraz Y z rozkładu N(mu,sigm). """
    X, Y = np.zeros(n), np.zeros(n)
    i = 0
    while i != n:
        V1, V2 = np.random.uniform(), np.random.uniform()
        X[i] = mu + sigm * np.sqrt(-2*np.log(V2))*np.cos(2*np.pi*V1)
        Y[i] = mu + sigm * np.sqrt(-2*np.log(V2))*np.sin(2*np.pi*V1)
        i += 1
    return X, Y
```

## Metoda akceptacji-odrzucenia przy użyciu rozkładu wykładniczego [2]



Jest to metoda korzystająca z gęstości standardowego rozkładu normalnego $f(x) = \frac{2}{\sqrt{2\pi}} \cdot e^{\frac{-x^2}{2}},$ gdzie $\: 0 < x < ∞ $ oraz gęstości rozkładu wykładniczego o intensywności $\lambda = 1$: $g(x) = e^{-x},$ gdzie $\: 0 < x < \infty $.\
Szukana stała c jest wyliczana poniższym sposobem: $$ c =  Max \frac{f(x)}{g(x)} = Max \sqrt{\frac{2}{\pi}} \cdot e^{x-\frac{x^2}{2}} = \frac{f(1)}{g(1)} = \sqrt{\frac{2e}{\pi}}. $$
Wtedy
$$ \frac{f(x)}{cg(x)} = e^{x-\frac{x^2}{2} - \frac{1}{2}} = e^{-\frac{(x-1)^2}{2}}.  $$

Korzystając z klasycznej wersji metody akceptacji-odrzucenia, algorytm wyglądałby wówczas tak:
1. Generuj $ Y \sim Exp(1)$.
2. Generuj $ U \sim U(0,1)$.
3. Jeśli $U \leq e^{-\frac{(Y-1)^2}{2}} $, wstaw $X=Y$. W przeciwnym razie wróć do 1. \


Wtedy wygenerowana jest wartość bezwzględna zmiennej losowej o standardowym rozkładzie normalnym. Przy założeniu, że pewna zmienna przyjmuje z takim samym prawdopodobieństwem $ X, \:$ lub $ -X$, otrzymuje się wtedy zmienną losową $Z$ o standardowym rozkładzie normalnym.\
Warunek w 3. kroku $U \leq e^{-\frac{(Y-1)^2}{2}} $ można równoważnie zapisać jako $ -log{U} \geq -\frac{(Y-1)^2}{2} $. Ponieważ jednak $-log{U} \sim Exp(1)$ , równoznacznie przyjmuje się:


1. Generuj $ Y_1 \sim Exp(1)$, $\: Y_2 \sim Exp(1)$.
2. Jeśli $Y_2 \geq -\frac{(Y_1-1)^2}{2} $, wstaw $X=Y_1$. W przeciwnym razie wróć do 1.


Wiadomo teraz, że  $Y_2 \geq -\frac{(Y_1-1)^2}{2} $. Ponieważ dane jest $ Y_2 \sim Exp(1)$, zatem różnica, o którą $Y_2$ przekracza wartość $-\frac{(Y_1-1)^2}{2} $, korzystając z własności braku pamięci, również ma rozkład wykładniczy o intensywności $1$. Oznacza to, że w kroku 2. jest generowana także zmienna losowa o rozkładzie wykładniczym i $\lambda = 1$, niezależna od X. Ostateczny zapis algorytmu widoczny jest w poniższym punkcie.


### Algorytm

1. Generuj $Y_1 \sim Exp(1), \:\: Y_2 \sim Exp(1), \:\:  Y_1 \: \perp \: Y_2. $
2. Jeśli $ Y_2 - \frac{(Y_1 - 1)^{2}}{2} > 0 $, wstaw $ Y = Y_2 - \frac{(Y_1 - 1)^{2}}{2} $. W przeciwnym razie wróć do 1.
3. Generuj $U \sim U(0,1)$, wstaw
$ Z = \begin{cases}Y_1, & U \leq \frac{1}{2}\\-Y_1, & U > \frac{1}{2}
\end{cases}$.


Wtedy $Z \sim N(0,1), \:\: Y \sim Exp(1), \:\: Z \perp Y$.

### Implementacja

```python
def elimination_gen(mu = 0, sigm = 1, n=1000):
    Y, Z = np.zeros(n), np.zeros(n)
    i = 0
    while i != n:
        Y1, Y2 = np.random.exponential(1, size=2)
        if Y2 - ((Y1 - 1)**2)/2 > 0:
            Y[i] = Y2 - ((Y1 - 1)**2)/2
        else:
            continue
        U = np.random.uniform(0, 1, size=1)
        if U <= 1/2:
            Z[i] = mu + sigm *Y1
        else:
            Z[i] = mu + sigm *(-Y1)
        i += 1
    return Z, Y
```

```python
import time
import numpy as np
n = 10000
num_of_samples = 100
def time_analysis(n, num_of_samples):
    boxmuller_time, polar_time, elimination_time = np.zeros(num_of_samples), np.zeros(num_of_samples), np.zeros(num_of_samples)

    for i in range(1, num_of_samples*n + 1, n):
        start_1 = time.time()
        boxmuller_gen(0, 1, i)[0]
        stop_1 = time.time()
        boxmuller_time[int((i-1)/n)] = stop_1 - start_1

        start_2 = time.time()
        polar_gen(0, 1, i)[0]
        stop_2 = time.time()
        polar_time[int((i-1)/n)] = stop_2 - start_2

        start_3 = time.time()
        elimination_gen(0, 1, i)[0]
        stop_3 = time.time()
        elimination_time[int((i-1)/n)] = stop_3 - start_3

    return boxmuller_time, polar_time, elimination_time

plt.plot(np.arange(1, num_of_samples*n + 1, n), time_analysis(n, num_of_samples)[0], color = 'b', label = "Box-Muller")
plt.plot(np.arange(1, num_of_samples*n + 1, n), time_analysis(n, num_of_samples)[1], color = 'r', label = "Biegunowa")
plt.plot(np.arange(1, num_of_samples*n + 1, n), time_analysis(n, num_of_samples)[2], color = 'g', label = "Akceptacji - odrzucenia")
plt.title("Porównanie czasów wykonania dla metod generowania r. normalnego")
plt.xlabel("Długość próbki")
plt.ylabel("Czas wykonania [s]")
plt.legend()
plt.show()
```

Może to wynikać z faktu, że metoda biegunowa wykorzystuje pośrednie obliczanie funkcji sinus oraz cosinus losowego kąta, gdzie metoda Boxa-Mullera przeprowadza takie obliczenia w sposób bezpośredni. # Metoda Ziggurat [3]

Algorytm Ziggurat to ostatni z przedstawionych algorytmów generowania zmiennych losowych. Nie powiela dwóch głównych problemów metody akceptacji-odrzucenia, tj. nie odrzuca dużej części próbek oraz nie oblicza wartości funkcji gęstości prawdopodobieństwa f(x) dla każdego punktu.
Ta metoda natomiast pokrywa prawą połowę wykresu gęstości rozkładu szeregiem poziomych prostokątów o tych samych polach, w celu pokrycia go możliwie najefektywniej, tj. minimalizuje obszar poza krzywą rozkładu.


image.png

Rys. 1 - Przykładowy diagram metody Ziggurata dla rozkładu normalnego. [3]

Przyjmując, że $0 = x_0 < x_1 < ... < x_{255} = r\:$ to punkty odpowiadające prawym krawędziom prostokątów. Każdy prostokąt ma przypisany indeks $R_i, i=0,1,2,...,255$, z uwagą, że prostokąt $R_0$ jest pusty, ponieważ jego prawa krawędź $x_0 = 0$ jest jednocześnie lewą krawędzią $R_1$. \
Ustalając konkretny prostokąt $R_{i}, \: i>0$, losowa współrzędna $x$ punktu w prostokącie $R_{i}$ jest równa $Ux_i$ dla $U \sim U(0,1)$. Jeśli również spełnia warunek $x < x_{i-1}$, to wiadomo, bez potrzeby generowania $y$, że punkt o współrzędnych $(x,y)$ znajduje się w obszarze ograniczonym krzywą gęstości.\
Natomiast wybierając $(x,y)$ z dolnego obszaru, tj. prostokąta połączonego w punkcie $x_{255} = r$ z ogonem rozkładu oraz $ v= rf(r) + \int_r^\infty f(x)\,\mathrm{d}x$, dany jest krótki algorytm:


1. Generuj $x= \frac{vU}{f(r)}, \:\: U \sim U(0,1)$.
2. Jeśli $x < r$, zwróć $x$. W przeciwnym razie zwróć $x$ z ogona rozkładu.

Wtedy, z prawdopodobieństwem równym $\frac{rf(r)}{v}$, zostaje wygenerowane $x$ z ustalonego prostokąta.\
Na podstawie powyższych, otrzymuje się:


$$[f(x_{i-1}) - f(x_i)] = v, \: \:\:\:\:  \: i=1,2,...,255.$$

Punkt $r$ wybierany jest za pomocą metody bisekcji, aby spełniona była nierówność $ |v- x_1 +x_1f(x_1)| < \varepsilon$, gdzie $\varepsilon$ jest pewną małą wartością oraz
$$ x_i = f^{-1}(\frac{v}{x_{i+1}} + f(x_{i+1})), \:\:\:\: i=254,253,...,1.$$

Aby przyspieszyć wykonywanie metody Ziggurat, do ostatecznego algorytmu zostało dodane generowanie $32$-bitowej liczby całkowitej używanej do tworzenia $x$:\
Dla każdego $i$ w $1 \leq i \leq 255$ podstawić $k_i = \lfloor 2^{32} (\frac{x_{i-1}}{x_i}) \rfloor $ oraz $w_i = 0.5^{32}x_i $, natomiast dla $i=0$ podstawić $k_0 = \lfloor 2^{32} (\frac{rf(r)}{v}) \rfloor $ oraz $ w_0 = 0.5^{32}\frac{v}{f(r)} $.






### Algorytm

1. Generuj $32$-bitową liczbę całkowitą $j$, a $i$ to indeks utworzony przez $8$ ostatnich bitów $j$.
2. Wstaw $x = jw_i$. Jeśli $j < k_i$, zwróć $x$.
3. Jeśli $i = 0$, generuj realizację $x$ z ogona rozkładu Pareto, za pomocą metody odwracania dystrybuanty.\
  3.1. Generuj $u \sim U(0,1)$.\
  3.2. Wstaw $u = 1 - (\frac{m}{x})^{k}$, gdzie $m$ to parametr skali, a $k$ to parametr kształtu rozkładu Pareto.\
  3.3. Wstaw $x = \frac{m}{(1-u)^{\frac{1}{k}}}$.
4. Jeśli $[f(x_{i-1}) - f(x_i)]U < f(x) - f(x_i)$, zwróć $x$. Tutaj $U \sim U(0,1).$
5. Wróć do punktu 1.

### Implementacja

```python
import dataclasses
import numpy as np
import numpy.typing
import random
import matplotlib.pyplot as plt
import scipy.stats as stats
import scipy.integrate
import math

from numpy.typing import ArrayLike as NumpyArray

two_to_32 = 2**32
half_to_32 = 0.5**32

pareto_k = 1
pareto_m = 1


@dataclasses.dataclass
class KWVectors:
    k: NumpyArray
    w: NumpyArray

@dataclasses.dataclass
class RVXVec:
    r: float
    v: float
    x_vec: NumpyArray


def generate_rvxvec(curve_function, reversed_curve_function):
    f = curve_function
    inv_f = reversed_curve_function

    r = 3.6541528853610088+2
    v = 0.00492867323399

    x_vec = numpy.zeros(256)
    x_vec[0] = pareto_m
    x_vec[255] = r

    for i in range(254, 0, -1):
        x_vec[i] = inv_f(v/x_vec[i + 1] + f(x_vec[i + 1]))
    return RVXVec(r, v, x_vec)


def generate_kw_vectors(x_vec, curve_function, r, v):
    k = numpy.zeros(256)
    w = numpy.zeros(256)

    f = curve_function

    k[0] = math.floor(two_to_32 * r * (f(r) / v))
    w[0] = half_to_32 * (v / f(r))

    for i in range(1, 256):
        k[i] = math.floor(two_to_32 * (x_vec[i - 1] / x_vec[i]))
        w[i] = half_to_32 * x_vec[i]

    return KWVectors(k, w)

def generate_single_ziggurat_value(x_vec, kw, curve_function):
    f = curve_function
    while True:
        j = random.getrandbits(32)
        i = j & 0x000000FF

        value = j * kw.w[i]
        if j < kw.k[i]:
            if value > pareto_m:
                return value
            continue

        if i == 0:
            u = np.random.rand()
            return (pareto_m * (1 - u)**(-1 / pareto_k))

        u = random.uniform(0, 1)
        if (f(x_vec[i - 1]) - f(x_vec[i])) * u < f(value) - f(x_vec[i]):
            return value



def generate_ziggurat_values(count, curve_function,reversed_curve_function):

    rvxvec = generate_rvxvec(curve_function, reversed_curve_function)
    kwvectors = generate_kw_vectors(rvxvec.x_vec, curve_function, rvxvec.r, rvxvec.v)

    output = np.zeros(count)
    for i in range(count):
        output[i] = generate_single_ziggurat_value(rvxvec.x_vec, kwvectors, curve_function)

    return output

def pareto_distribution_density(x):
    return (pareto_k * (pareto_m**pareto_k)) / x**(pareto_k + 1)

def reversed_pareto_distribution_density(y):
    return (pareto_k * (pareto_m**pareto_k / y ))**(1/(pareto_k + 1))

```
